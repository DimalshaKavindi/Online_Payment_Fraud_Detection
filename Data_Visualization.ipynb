```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv(r'onlinefraud.csv')
payment_data= data.copy()
```

```python
payment_data['Hour'] = (payment_data['step']-1)%24
sns.countplot(payment_data, x= 'Hour')
```

```python
payment_type_count = payment_data['type'].value_counts()
payment_percentage = (payment_type_count / payment_type_count.sum()) * 100

plt.figure(figsize=(8, 6))
payment_percentage.plot(kind='bar')  # Use the plot method for Series objects

plt.xlabel('Payment Method')
plt.ylabel("Percentage")
plt.title("Percentage of Different Payment Method")

plt.show()
```

```python
fraud_data = payment_data[payment_data['isFraud'] == 1]
payment_type_count_fraud = fraud_data['type'].value_counts()
payment_type_percentage_fraud = (payment_type_count_fraud / payment_type_count_fraud.sum()) * 100

print(payment_type_percentage_fraud)

plt.figure(figsize=(8, 6))
payment_type_percentage_fraud.plot(kind='bar')  # Use the plot method for Series objects

plt.xlabel('Payment Method')
plt.ylabel("Percentage")
plt.title("Percentage of Different Fraud Payment Method")

plt.show()

```
